```python
##make sure to pip install networkx before running
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/elizabethterveen/Downloads/twitch/ENGB/musae_ENGB_edges.csv")
```

```python
##simple graph visualization 
G = nx.from_pandas_edgelist(df,source='from',target='to')
edges = nx.draw_networkx_edges(G, pos=nx.spring_layout(G),node_size = 100,width =0.2)
plt.show()
```

```python
plt.figure(num=None, figsize=(20, 20), dpi=80)
plt.axis('off')
fig = plt.figure(1)
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos)
nx.draw_networkx_edges(G,pos)
nx.draw_networkx_labels(G,pos)

cut = 1.00
xmax = cut * max(xx for xx, yy in pos.values())
ymax = cut * max(yy for xx, yy in pos.values())
plt.xlim(0, xmax)
plt.ylim(0, ymax)

plt.show()
```
